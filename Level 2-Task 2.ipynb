```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
```

```python
dataset=pd.read_csv("Dataset_for_Internship.csv",encoding='unicode_escape')
dataset.head(5)
```

# Task 2:Cuisine Combination

### 1. Identify the most common combinations of cuisines in the dataset.

```python
#Filter out rows where the 'Cuisines' column contains multiple cuisines (separated by commas)
multi_cuisine_entries = dataset[dataset['Cuisines'].str.contains(',')]

#Split the cuisines by commas and convert to tuples
cuisine_combinations = multi_cuisine_entries['Cuisines'].str.split(',').apply(lambda x: tuple(sorted(x)))

#Count the occurrences of each cuisine combination
most_common_combination = Counter(cuisine_combinations).most_common()

#Convert the result into a DataFrame
common_combination_df = pd.DataFrame(most_common_combination, columns=['Cuisine Combination', 'Count'])
common_combination_df.head(11)
```

This data highlights the popularity of cuisine combinations in a restaurant dataset. The most popular combination is "Chinese and North Indian," with 511 occurrences, followed by "Mughlai and North Indian" with 334. These combinations suggest a strong preference for North Indian cuisine paired with other popular types like Chinese and Mughlai. The data indicates that fusion or multi-cuisine offerings tend to attract more customers.

### 2. Determine if certain cuisine combinations tend to have higher ratings

```python
#Filter out rows where the 'Cuisines' column contains multiple cuisines (separated by commas)
multi_cuisine_entries = dataset[dataset['Cuisines'].str.contains(',')].copy()  

#Split the cuisines by commas, sort them, and convert to tuples
multi_cuisine_entries.loc[:, 'Cuisine Combination'] = multi_cuisine_entries['Cuisines'].str.split(',').apply(lambda x: tuple(sorted(x)))

#Group by the 'Cuisine Combination' and calculate the average rating
avg_rating_by_combination = multi_cuisine_entries.groupby('Cuisine Combination')['Aggregate rating'].mean().reset_index()

#Sort by average rating in descending order to see which combinations tend to have higher ratings
avg_rating_by_combination_sorted = avg_rating_by_combination.sort_values(by='Aggregate rating', ascending=False)
avg_rating_by_combination_sorted.head(70)
```

Popular cuisine combinations like "Chinese, Mughlai, North Indian" and "Chinese, Mughlai, North Indian, Biryani" tend to have high ratings around 4.6, indicating a positive correlation between popularity and good customer satisfaction.
However, some of the highest-rated combinations (e.g., "Caribbean, Seafood, American" and "Lucknowi, Mughlai") are less common but still achieve top ratings of 4.9. This suggests that niche or less frequent combinations can still achieve high customer satisfaction.
